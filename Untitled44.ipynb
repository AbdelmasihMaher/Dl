```python
# 1. Import Libraries
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout,
    BatchNormalization
)

from tensorflow.keras.utils import to_categorical
```

```python
# 2. Data Requirements & Preprocessing

# Load Dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalization
X_train = X_train / 255.0
X_test = X_test / 255.0

# Reshape for CNN
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# One Hot Encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 3. Model Architecture Function

def build_cnn_model():
    model = Sequential()

    # First Convolution Layer
    model.add(Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2,2)))

    # Second Convolution Layer
    model.add(Conv2D(64, (3,3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))

    # Flatten Layer
    model.add(Flatten())

    # Dense Layer
    model.add(Dense(128, activation='relu'))

    # Dropout Layer (Enhancement Requirement)
    model.add(Dropout(0.5))

    # Output Layer
    model.add(Dense(10, activation='softmax'))

    return model
```

```python
# 7. Experimentation Requirements (Mandatory)

# Experiment 1: Adam Optimizer (Model A)
print("\n=== Starting Experiment 1: Adam Optimizer ===")
model_adam = build_cnn_model()
model_adam.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_adam = model_adam.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)

test_loss_adam, test_accuracy_adam = model_adam.evaluate(X_test, y_test, verbose=0)


# Experiment 2: SGD Optimizer (Model B)
print("\n=== Starting Experiment 2: SGD Optimizer ===")
model_sgd = build_cnn_model()
model_sgd.compile(
    optimizer='sgd',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_sgd = model_sgd.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)

test_loss_sgd, test_accuracy_sgd = model_sgd.evaluate(X_test, y_test, verbose=0)
```

```python
# 8. Visualization Requirements

def plot_experiment_curves(history, title):
    # Training vs. Validation Accuracy curves
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'{title} - Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'])

    # Training vs. Validation Loss curves
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'{title} - Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'])
    plt.show()

# Plot for both experiments
plot_experiment_curves(history_adam, "Model A (Adam)")
plot_experiment_curves(history_sgd, "Model B (SGD)")

```

```python
# 9. Results Requirements (Comparison Table)
# =========================================

print("\n==================================================")
print(f"{'Model':<20} | {'Accuracy':<15} | {'Loss':<10}")
print("--------------------------------------------------")
print(f"{'Model A (Adam)':<20} | {test_accuracy_adam*100:.2f}%{'':<9} | {test_loss_adam:.4f}")
print(f"{'Model B (SGD)':<20} | {test_accuracy_sgd*100:.2f}%{'':<9} | {test_loss_sgd:.4f}")
print("==================================================")
```
